````
from keras import models
from keras import layers

# ëª¨ë¸ì˜ ë ˆì´ì–´ ì¶”ê°€
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,),name='Input_Layer'))
network.add(layers.Dense(10, activation='softmax',name='Output_Layer'))

from keras.utils import plot_model
plot_model(network, show_shapes=True)
````

ì‹ ê²½ë§ì„ í•™ìŠµì‹œí‚¤ê¸° ìœ„í•´, 
* optimizer 
* loss function 
* metric

ì„ ì •ì˜í•˜ì˜€ìŠµë‹ˆë‹¤ 

````
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
````

##### **ëª¨ë¸ í•™ìŠµ**
ì•ì—ì„œ ì‚´í´ë³¸ ê²ƒê³¼ ê°™ì´ ìš°ë¦¬ì˜ í›ˆë ¨ ì´ë¯¸ì§€ëŠ” [0, 255] ì‚¬ì´ì˜ ê°’ì¸ uint8 íƒ€ì…ì˜ (60000, 28, 28) í¬ê¸°ë¥¼ ê°€ì§„ ë°°ì—´ë¡œ ì €ì¥ë˜ì–´ ìˆìŠµë‹ˆë‹¤.


dense layerì— ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ ì…ë ¥í•˜ê¸° ìœ„í•˜ì—¬ 
28 x 28 ì˜ 2ì°¨ì› ë°ì´í„°ë¥¼ 28*28 ê¸¸ì´ë¥¼ ê°–ëŠ” 1ì°¨ì› ë°ì´í„°ì˜ í˜•íƒœë¡œ ë°”ê¾¸ì—ˆìŠµë‹ˆë‹¤. 


ê·¸ë¦¬ê³  ê°’ì˜ ë²”ìœ„ë¥¼ 0ê³¼ 1ì‚¬ì´ì˜ float32 í˜•íƒœë¡œ ìˆ˜ì •í•˜ì˜€ìŠµë‹ˆë‹¤. 


ë§ˆì§€ë§‰ìœ¼ë¡œ í•™ìŠµë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì˜ ë ˆì´ë¸”ì„ ëª¨ë¸ì˜ ì•„ì›ƒí’‹ nodeì— ì í•©í•˜ë„ë¡ ë²”ì£¼í˜•(categorical)ìœ¼ë¡œ ë‹¤ì‹œ í‘œí˜„í•˜ì˜€ìŠµë‹ˆë‹¤. 

````
# í•™ìŠµ ë°ì´í„°ì˜ í‘œí˜„ í˜•íƒœ ë³€ê²½ (2d array --> 1d array)
input_train_images = train_images.reshape((60000, 28 * 28))
# í•™ìŠµ ë°ì´í„°ì˜ ê°’ì˜ ë²”ì£¼ ë³€ê²½ (0~255 --> 0~1)
input_train_images = input_train_images.astype('float32') / 255

input_test_images = test_images.reshape((10000, 28 * 28))
input_test_images = input_test_images.astype('float32') / 255


from keras.utils import to_categorical

# ë ˆì´ë¸”ì„ ë²”ì£¼í˜•(one-hot)ìœ¼ë¡œ ë³€ê²½ 
onehot_train_labels = to_categorical(train_labels)
onehot_test_labels = to_categorical(test_labels)

print ("í•™ìŠµ ë°ì´í„°ì˜ 0ë²ˆ ë ˆì´ë¸” ì¶œë ¥:", train_labels[0])
print ("í•™ìŠµ ë°ì´í„°ì˜ 0ë²ˆ ë ˆì´ë¸”ì„ ë²”ì£¼í˜•ìœ¼ë¡œ ì¶œë ¥:",onehot_train_labels[0])
````

````
network.fit(input_train_images, 
            onehot_train_labels, 
            epochs=5, 
            batch_size=128)
````

##### **ëª¨ë¸ í…ŒìŠ¤íŠ¸**

ê°„ë‹¨í•œ ëª¨ë¸ì˜ í•™ìŠµì´ ì™„ë£Œë˜ì—ˆìœ¼ë‹ˆ, ì„±ëŠ¥ í…ŒìŠ¤íŠ¸ë¥¼ ì§„í–‰í•˜ì—¬ ë³´ê² ìŠµë‹ˆë‹¤. 

````
# í…ŒìŠ¤íŠ¸ì…‹ í…ŒìŠ¤íŠ¸ 
test_loss, test_acc = network.evaluate(input_test_images, onehot_test_labels)

#í…ŒìŠ¤íŠ¸ ê²°ê³¼ ì¶œë ¥
print('test_acc:', test_acc)
````

### **3) ë‹¤ë¥¸ ëª¨ë¸ë„ ë‹¤ë£¨ê¸°** (e.g., CNN)

> ì´ ë…¸íŠ¸ë¶ì€ [ì¼€ë¼ìŠ¤ ì°½ì‹œìì—ê²Œ ë°°ìš°ëŠ” ë”¥ëŸ¬ë‹](https://tensorflow.blog/%EC%BC%80%EB%9D%BC%EC%8A%A4-%EB%94%A5%EB%9F%AC%EB%8B%9D/) ì±…ì˜ 5ì¥ 1ì ˆì˜ ì½”ë“œ ì˜ˆì œì…ë‹ˆë‹¤. ì±…ì—ëŠ” ë” ë§ì€ ë‚´ìš©ê³¼ ê·¸ë¦¼ì´ ìˆìŠµë‹ˆë‹¤. ì´ ë…¸íŠ¸ë¶ì—ëŠ” ì†ŒìŠ¤ ì½”ë“œì— ê´€ë ¨ëœ ì„¤ëª…ë§Œ í¬í•¨í•©ë‹ˆë‹¤.


ì»¨ë¸Œë„·ì˜ ì •ì˜ì™€ ì»¨ë¸Œë„·ì´ ì»´í“¨í„° ë¹„ì „ ê´€ë ¨ ì‘ì—…ì— ì˜ ë§ëŠ” ì´ìœ ì— ëŒ€í•´ ì´ë¡ ì  ë°°ê²½ì„ ì•Œì•„ë³´ê² ìŠµë‹ˆë‹¤. í•˜ì§€ë§Œ ë¨¼ì € ê°„ë‹¨í•œ ì»¨ë¸Œë„· ì˜ˆì œë¥¼ ë‘˜ëŸ¬ ë³´ì£ . 2ì¥ì—ì„œ ì™„ì „ ì—°ê²° ë„¤íŠ¸ì›Œí¬ë¡œ í’€ì—ˆë˜ MNIST ìˆ«ì ì´ë¯¸ì§€ ë¶„ë¥˜ì— ì»¨ë¸Œë„·ì„ ì‚¬ìš©í•´ ë³´ê² ìŠµë‹ˆë‹¤. ê¸°ë³¸ì ì¸ ì»¨ë¸Œë„·ì´ë”ë¼ë„ 2ì¥ì˜ ì™„ì „ ì—°ê²°ëœ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í›¨ì”¬ ì•ì§€ë¥¼ ê²ƒì…ë‹ˆë‹¤.

ë‹¤ìŒ ì½”ë“œëŠ” ê¸°ë³¸ì ì¸ ì»¨ë¸Œë„·ì˜ ëª¨ìŠµì…ë‹ˆë‹¤. `Conv2D`ì™€ `MaxPooling2D` ì¸µì„ ìŒ“ì•„ ì˜¬ë ¸ìŠµë‹ˆë‹¤. ì ì‹œ í›„ì— ì´ë“¤ì´ ë¬´ì—‡ì¸ì§€ ë°°ìš°ê² ìŠµë‹ˆë‹¤.

ì»¨ë¸Œë„·ì´ `(image_height, image_width, image_channels)` í¬ê¸°ì˜ ì…ë ¥ í…ì„œë¥¼ ì‚¬ìš©í•œë‹¤ëŠ” ì ì´ ì¤‘ìš”í•©ë‹ˆë‹¤(ë°°ì¹˜ ì°¨ì›ì€ í¬í•¨í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤). ì´ ì˜ˆì œì—ì„œëŠ” MNIST ì´ë¯¸ì§€ í¬ë§·ì¸ `(28, 28, 1)` í¬ê¸°ì˜ ì…ë ¥ì„ ì²˜ë¦¬í•˜ë„ë¡ ì»¨ë¸Œë„·ì„ ì„¤ì •í•´ì•¼ í•©ë‹ˆë‹¤. ì´ ë•Œë¬¸ì— ì²« ë²ˆì§¸ ì¸µì˜ ë§¤ê°œë³€ìˆ˜ë¡œ `input_shape=(28, 28, 1)`ì„ ì „ë‹¬í•©ë‹ˆë‹¤.

````
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
````

ì§€ê¸ˆê¹Œì§€ ì»¨ë¸Œë„· êµ¬ì¡°ë¥¼ ì¶œë ¥í•´ ë³´ì£ :

````
model.summary()
````

`Conv2D`ì™€ `MaxPooling2D` ì¸µì˜ ì¶œë ¥ì€ `(height, width, channels)` í¬ê¸°ì˜ 3D í…ì„œì…ë‹ˆë‹¤. ë†’ì´ì™€ ë„“ì´ ì°¨ì›ì€ ë„¤íŠ¸ì›Œí¬ê°€ ê¹Šì–´ì§ˆìˆ˜ë¡ ì‘ì•„ì§€ëŠ” ê²½í–¥ì´ ìˆìŠµë‹ˆë‹¤. ì±„ë„ì˜ ìˆ˜ëŠ” `Conv2D` ì¸µì— ì „ë‹¬ëœ ì²« ë²ˆì§¸ ë§¤ê°œë³€ìˆ˜ì— ì˜í•´ ì¡°ì ˆë©ë‹ˆë‹¤(32ê°œ ë˜ëŠ” 64ê°œ).

ë‹¤ìŒ ë‹¨ê³„ì—ì„œ ë§ˆì§€ë§‰ ì¸µì˜ (`(3, 3, 64)` í¬ê¸°ì¸) ì¶œë ¥ í…ì„œë¥¼ ì™„ì „ ì—°ê²° ë„¤íŠ¸ì›Œí¬ì— ì£¼ì…í•©ë‹ˆë‹¤. ì´ ë„¤íŠ¸ì›Œí¬ëŠ” ì´ë¯¸ ìµìˆ™í•˜ê²Œ ë³´ì•˜ë˜ `Dense` ì¸µì„ ìŒ“ì€ ë¶„ë¥˜ê¸°ì…ë‹ˆë‹¤. ì´ ë¶„ë¥˜ê¸°ëŠ” 1D ë²¡í„°ë¥¼ ì²˜ë¦¬í•˜ëŠ”ë° ì´ì „ ì¸µì˜ ì¶œë ¥ì´ 3D í…ì„œì…ë‹ˆë‹¤. ê·¸ë˜ì„œ ë¨¼ì € 3D ì¶œë ¥ì„ 1D í…ì„œë¡œ í¼ì³ì•¼ í•©ë‹ˆë‹¤. ê·¸ë‹¤ìŒ ëª‡ ê°œì˜ `Dense` ì¸µì„ ì¶”ê°€í•©ë‹ˆë‹¤:

````
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
````

10ê°œì˜ í´ë˜ìŠ¤ë¥¼ ë¶„ë¥˜í•˜ê¸° ìœ„í•´ ë§ˆì§€ë§‰ ì¸µì˜ ì¶œë ¥ í¬ê¸°ë¥¼ 10ìœ¼ë¡œ í•˜ê³  ì†Œí”„íŠ¸ë§¥ìŠ¤ í™œì„±í™” í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤. ì´ì œ ì „ì²´ ë„¤íŠ¸ì›Œí¬ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

````
model.summary()
````

ì—¬ê¸°ì—ì„œ ë³¼ ìˆ˜ ìˆë“¯ì´ `(3, 3, 64)` ì¶œë ¥ì´ `(576,)` í¬ê¸°ì˜ ë²¡í„°ë¡œ í¼ì³ì§„ í›„ `Dense` ì¸µìœ¼ë¡œ ì£¼ì…ë˜ì—ˆìŠµë‹ˆë‹¤.

ì´ì œ MNIST ìˆ«ì ì´ë¯¸ì§€ì— ì´ ì»¨ë¸Œë„·ì„ í›ˆë ¨í•©ë‹ˆë‹¤. 2ì¥ì˜ MNIST ì˜ˆì œ ì½”ë“œë¥¼ ë§ì´ ì¬ì‚¬ìš©í•˜ê² ìŠµë‹ˆë‹¤.

````
from keras.datasets import mnist
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

input_train_images = train_images.reshape((60000, 28, 28, 1))
input_train_images = input_train_images.astype('float32') / 255

input_test_images = test_images.reshape((10000, 28, 28, 1))
input_test_images = input_test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
````

````
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(input_train_images, train_labels, epochs=5, batch_size=64)
````

í…ŒìŠ¤íŠ¸ ë°ì´í„°ì—ì„œ ëª¨ë¸ì„ í‰ê°€í•´ ë³´ì£ :

````
test_loss, test_acc = model.evaluate(input_test_images, test_labels)
````
